```python
eda_df['roi_manual'] = (eda_df['gross'] - eda_df['budget']) / eda_df['budget']

eda_df['roi_manual']
```

### Target Variable Exploration

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_features = ['imdb_score', 'roi_manual', 'budget', 'gross', 'vote_average', 'vote_count', 'popularity']

for col in numerical_features:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    sns.histplot(data=eda_df, x=col, kde=True, ax=axes[0])
    axes[0].set_title(f'Histogram of {col}')

    sns.boxplot(data=eda_df, x=col, ax=axes[1])
    axes[1].set_title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Filter and transform the data
filtered_data = eda_df[eda_df["roi_manual"] < 3000].copy()
filtered_data["log_roi_manual"] = np.log10(filtered_data["roi_manual"])

# Plot
sns.histplot(data=filtered_data, x="log_roi_manual")
plt.xlabel("log10(roi_manual)")
plt.show()
```

```python
categorical_features = ['genre', 'company', 'writer', 'director']

for col in categorical_features:
    plt.figure(figsize=(10, 4))
    eda_df[col].value_counts(normalize=True).head(10).plot(kind='bar')
    plt.title(f'Distribution of Top 10 {col}')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14,5))
sns.histplot(eda_df['imdb_score'], kde=True, ax=axes[0])
axes[0].set_title('IMDB Score Distribution')

sns.histplot(eda_df['roi_manual'], kde=True, ax=axes[1])
axes[1].set_title('ROI Distribution')
```

```python
correlation = eda_df.corr(numeric_only=True)
plt.figure(figsize=(12,10))
sns.heatmap(correlation, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Matrix')
```
