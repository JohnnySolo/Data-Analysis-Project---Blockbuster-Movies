## 1st Dataset: Movie Franchises



### Acquisition

```python
# 1. Movie Data Analysis Dataset

!wget https://raw.githubusercontent.com/JohnnySolo/Data-Analysis-Project---Blockbuster-Movies/main/movie.csv -O movie.csv

# Load the CSV file
import pandas as pd
movie_franchises = pd.read_csv("movie.csv")
movie_franchises.head()
```

### Preprocessing

```python
# Display data types and NA counts for each column
for column in movie_franchises.columns:
    print(f"Column: {column}")
    print(f"  Data Type: {movie_franchises[column].dtype}")
    print(f"  NA Count: {movie_franchises[column].isnull().sum()}")
    print("-" * 20)
```

```python
# Numerical Conversion (int, float)
movie_franchises['year'] = pd.to_numeric(movie_franchises['year'], errors='coerce')
movie_franchises['score'] = pd.to_numeric(movie_franchises['score'], errors='coerce')
movie_franchises['votes'] = pd.to_numeric(movie_franchises['votes'], errors='coerce')
movie_franchises['budget'] = pd.to_numeric(movie_franchises['budget'], errors='coerce')
movie_franchises['gross'] = pd.to_numeric(movie_franchises['gross'], errors='coerce')
movie_franchises['runtime'] = pd.to_numeric(movie_franchises['runtime'], errors='coerce')
```

```python
# Convert to categorical
movie_franchises['rating'] = movie_franchises['rating'].astype('category')
movie_franchises['genre'] = movie_franchises['genre'].astype('category')

# Create a mapping
unique_rating = movie_franchises['rating'].unique() # Get unique rating values
rating_mapping = {rating: i for i, rating in enumerate(unique_rating)} # Create a mapping based on the unique values

unique_genres = movie_franchises['genre'].unique() # Get unique genre values
genre_mapping = {genre: i for i, genre in enumerate(unique_genres)} # Create a mapping based on the unique values

# Apply the mapping to create *new* columns with numerical codes
movie_franchises['rating_code'] = movie_franchises['rating'].map(rating_mapping)
movie_franchises['genre_code'] = movie_franchises['genre'].map(genre_mapping)
```

```python
# Convert to categorical
for col in ['name', 'director', 'writer', 'star', 'country', 'country']:  # Variables that doesn't has a defined mapping
    movie_franchises[col] = movie_franchises[col].astype('category')
```

```python
# Convert to datetime by two steps:

## Convert released to two seperate columns (released_date and released_country)
movie_franchises['released_country'] = movie_franchises['released'].str.extract(r'\((.*)\)', expand=False)  # Extract country
movie_franchises['released_date'] = movie_franchises['released'].str.replace(r'\(.*\)', '', regex=True).str.strip()  # Remove country info

## 2nd step: convert the released_date column to a datetime type column based on various date structures
### Assumption: 'released_date' contains strings like 'November 1994', '1995', '2019', nan

import re

def parse_date(date_str):
    # Check if input is float and convert to string
    if isinstance(date_str, float):
        date_str = str(date_str)
    # Try parsing with format 'Month Day, Year'
    match = re.search(r'([a-zA-Z]+)\s+(\d{1,2}),\s+(\d{4})', date_str)
    if match:
        return pd.to_datetime(date_str, format='%B %d, %Y', errors='coerce')

    # Try parsing with format 'Month Year'
    match = re.search(r'([a-zA-Z]+)\s+(\d{4})', date_str)
    if match:
        return pd.to_datetime(date_str, format='%B %Y', errors='coerce')

    # Try parsing with format 'Year'
    match = re.search(r'(\d{4})', date_str)
    if match:
        return pd.to_datetime(date_str, format='%Y', errors='coerce')

    # If none of the above formats match, return NaT
    return pd.NaT

# Apply the parse_date function to the 'released_date' column
movie_franchises['released_date'] = movie_franchises['released_date'].apply(parse_date)
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = movie_franchises.isnull().sum() / len(movie_franchises) * 100

print(missing_percentage)
```

**Strategy for dealing with missing values - Imputation**

`SimpleImputer` class (from `Sklearn` library) provides basic strategies for imputing missing values. Missing values can be imputed with a provided constant value, or using the statistics (mean, median or most frequent) of each column in which the missing values are located.

<u>My strategy</u> - Due to the variables structure, I'll impute missing values in numerical columns and omit raws with missing categorical values.

```python
import pandas as pd
from sklearn.impute import SimpleImputer

# Identify numerical and categorical columns
numerical_cols = movie_franchises.select_dtypes(include=['number']).columns
categorical_cols = movie_franchises.select_dtypes(include=['object', 'category']).columns

# Create an imputer object with median strategy
imputer = SimpleImputer(strategy='median')

# Fit the imputer to the numerical data and transform
movie_franchises[numerical_cols] = imputer.fit_transform(movie_franchises[numerical_cols])

# Remove rows with missing values in categorical columns
movie_franchises.dropna(subset=categorical_cols, inplace=True)
```

Notes so far:
1. Omit NA's at first (besides budget)
2. Optional - replace budget if there's a better budget column in another table with less NA's

### Feature Engineering

We'll try to create new features that are likely to be relevant for predicting movie success or understanding factors that contribute to box office performance.

I added 4 new features (columns) to our table:

* `Profit`: Movie earnings (subtraction of 'budget' from 'gross')
* `ROI (Return on Investment)`: An index created by dividing 'profit' by 'budget'.
* `Release Season`: Extracting the season of the release (e.g., summer, winter) from the 'released' column.
* `Director/Writer/Star Popularity`: Using external data source (IMDB) to determine the popularity or success rate of directors, writers, and stars.

```python
# Create profit and roi columns

## Calculate profit
movie_franchises['profit'] = movie_franchises['gross'] - movie_franchises['budget']

## Calculate ROI
movie_franchises['roi'] = movie_franchises['profit'] / movie_franchises['budget']
```

```python
# Create release_season column

## Function to extract season from release date
def get_release_season(date):
    month = date.month
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Fall'
    else:
        return 'Winter'

## Apply the function to the 'released_date' column
movie_franchises['release_season'] = movie_franchises['released_date'].apply(get_release_season)
```

```python
# Create director_rating column

## Calculate average rating for each director, retaining current behavior
director_avg_ratings = movie_franchises.groupby('director', observed=False)['score'].mean()

## Merge with original DataFrame using 'transform' for efficient in-place update
movie_franchises['director_avg_rating'] = movie_franchises['director'].transform(lambda x: director_avg_ratings.get(x, float('nan')))
```

```python
# Final editing - Columns new order and new names

import pandas as pd

# Define the desired column order and new names
desired_column_order = {
    'name': 'title',
    'rating': 'rating_category',
    'genre': 'genre',
    'year': 'year',
    'released': 'released',
    'score': 'fans_average',
    'votes': 'fans_count' ,
    'director': 'director',
    'writer': 'writer',
    'star': 'star',
    'country': 'production_countries',
    'budget': 'budget',
    'gross': 'revenue',
    'company': 'production_companies',
    'runtime': 'runtime',
    'rating_code': 'rating_code',
    'genre_code': 'genre_code',
    'released_country': 'released_country',
    'released_date': 'released_date',
    'profit': 'profit',
    'roi': 'roi',
    'release_season': 'release_season',
    'director_avg_rating': 'director_avg_rating'
}

# Reorder the columns using only the keys of the dictionary
movie_franchises = movie_franchises[[col for col in desired_column_order.keys()]]
movie_franchises = movie_franchises.rename(columns=desired_column_order)

# Print the updated DataFrame's columns
print(movie_franchises.columns)
```

### Addition to Database

```python
import sqlite3
import pandas as pd

# Create a SQLite database
conn = sqlite3.connect('movie_database.db')
cursor = conn.cursor()

# Create the movie table
cursor.execute('''
CREATE TABLE IF NOT EXISTS movie_franchises (
    movie_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    rating_category FLOAT,
    genre TEXT,
    year INTEGER,
    released TEXT,
    fans_average FLOAT,
    fans_count INTEGER,
    director TEXT,
    writer TEXT,
    star TEXT,
    production_countries TEXT,  -- Changed from 'country'
    budget INTEGER,
    revenue INTEGER,  -- Changed from 'gross'
    production_companies TEXT,  -- Changed from 'company'
    runtime INTEGER,
    rating_code INTEGER, -- New column
    genre_code INTEGER, -- New column
    released_country TEXT, -- New column
    released_date TEXT, -- New column (store as TEXT and convert to datetime in pandas)
    profit INTEGER, -- New column
    roi REAL, -- New column
    release_season TEXT, -- New column
    director_avg_rating REAL -- New column
)
''')

# Insert data into the movie table
movie_franchises.to_sql('movie_franchises', conn, if_exists='replace', index=False)

# Commit the changes and close the connection
conn.commit()
conn.close()
```

### Normalization Steps for the Database

Normalization is a database design process that organizes data to reduce redundancy and improve integrity.
In our movie database project, normalization was applied to the `movie_metadata` table by separating genre and cast information into distinct tables.

---

**Normalization Rationale**

**1. First Normal Form (1NF)**  
To achieve 1NF, we'll ensure that each column contains atomic values. As of now, the columns for genres and cast contain multiple values within a single record, which violates 1NF. The solution will be to create separate tables for:  

- **`movie_genres`**: Stores the relationship between movies and their genres.  
- **`movie_cast`**: Stores the relationship between movies and their actors.

Before normalization, the `movie_metadata` table contained multiple values within a single column, such as:

| movie_id | title     | genre               | cast                    |
|----------|-----------|--------------------|-------------------------|
| 1        | Inception | Sci-Fi, Action      | Leonardo, Joseph, Ellen |
| 2        | Titanic   | Romance, Drama      | Leonardo, Kate          |

*Issues:*  
- `genre` and `cast` contain multiple values, violating atomicity (1NF).  

**Solution:** Split into separate tables.

| movie_id | genre     |
|----------|-----------|
| 1        | Sci-Fi     |
| 1        | Action     |
| 2        | Romance    |
| 2        | Drama      |

| movie_id | actor      |
|----------|------------|
| 1        | Leonardo   |
| 1        | Joseph     |
| 1        | Ellen      |
| 2        | Leonardo   |
| 2        | Kate       |

---

```python
!apt-get install sqlite3
```

```python
import sqlite3

# Create the first table
conn = sqlite3.connect('movies.db')
cursor = conn.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS movie_genres (
        movie_id INTEGER,
        genre TEXT,
        FOREIGN KEY (movie_id) REFERENCES movie_metadata(movie_id)
    );
''')
conn.commit()
conn.close()

# Create the second table after reopening connection
conn = sqlite3.connect('movies.db')
cursor = conn.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS movie_cast (
        movie_id INTEGER,
        actor_name TEXT,
        FOREIGN KEY (movie_id) REFERENCES movie_metadata(movie_id)
    );
''')
conn.commit()
conn.close()
```

```python
# Checking if our seperate tables were created
conn = sqlite3.connect('movies.db')
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
print(cursor.fetchall())  # This should list the created tables
conn.close()
```

**2. Second Normal Form (2NF)**  
2NF eliminates partial dependency by ensuring that non-key attributes are fully dependent on the primary key. Since genres and actors depended on a non-key attribute (movie ID), moving them to separate tables ensures that all attributes in each table are dependent on the full primary key.

Before normalization:

| movie_id | title     | franchise | genre   |
|----------|-----------|------------|---------|
| 1        | Inception | Nolan Films | Sci-Fi  |
| 2        | Batman    | Nolan Films | Action  |

*Issue:*  
- `franchise` depends on `movie_id` but should depend on an independent `franchise_id`.

**Solution:** Separate into:

| franchise_id | franchise   |
|--------------|-------------|
| 1            | Nolan Films  |

| movie_id | title     | franchise_id |
|----------|-----------|--------------|
| 1        | Inception | 1            |
| 2        | Batman    | 1            |

---

```python
import sqlite3

# Connect to the database
conn = sqlite3.connect('movies.db')
cursor = conn.cursor()

# Create a new table for movie runtime, removing dependency on the full primary key
cursor.execute('''
    CREATE TABLE IF NOT EXISTS movie_runtime (
        movie_id INTEGER,
        runtime INTEGER,
        FOREIGN KEY (movie_id) REFERENCES movie_metadata(movie_id)
    );
''')

# Create a new table for movie ratings, removing dependency on the full primary key
cursor.execute('''
    CREATE TABLE IF NOT EXISTS movie_ratings (
        movie_id INTEGER,
        rating FLOAT,
        FOREIGN KEY (movie_id) REFERENCES movie_metadata(movie_id)
    );
''')

conn.commit()
conn.close()
```

**3. Third Normal Form (3NF)**  
3NF eliminates transitive dependencies. By moving genre and actor attributes to separate tables, we ensure that attributes are directly related to the movie ID without intermediary dependencies.

Before normalization:

| movie_id | budget  | revenue  | profit  |
|----------|---------|---------|---------|
| 1        | 1000000 | 5000000  | 4000000 |

*Issue:*  
- `profit` can be derived from `revenue - budget`, making it a derived attribute.

**Solution:** Remove `profit` and calculate dynamically.

---

**Updated Database Schema**

After normalization, the following schema will be adopted:

- **`movie_metadata`**: (movie_id, title, runtime, release_year, rating)  
- **`movie_genres`**: (movie_id, genre)  
- **`movie_cast`**: (movie_id, actor_name)  

---


```python
conn = sqlite3.connect('movies.db')
cursor = conn.cursor()

# Create a separate table for production companies
cursor.execute('''
    CREATE TABLE IF NOT EXISTS production_companies (
        company_id INTEGER PRIMARY KEY AUTOINCREMENT,
        company_name TEXT UNIQUE
    );
''')

# Link production companies to movies via foreign key
cursor.execute('''
    CREATE TABLE IF NOT EXISTS movie_production (
        movie_id INTEGER,
        company_id INTEGER,
        FOREIGN KEY (movie_id) REFERENCES movie_metadata(movie_id),
        FOREIGN KEY (company_id) REFERENCES production_companies(company_id)
    );
''')

conn.commit()
conn.close()
```

**Conclusion**  

By applying normalization techniques, the database will:

- Reduce redundancy and improve data integrity.  
- Support efficient querying and maintenance.  
- Handle many-to-many relationships effectively.

### Insert Data into SQLite Database

Based on the normalization we did, we'll now form our DB for the project.

```python
import pandas as pd

# Load CSV files to inspect the data
metadata_df = pd.read_csv('movie_metadata.csv')
genres_df = pd.read_csv('movie_franchises.csv')
quotes_df = pd.read_csv('movie_quotes.csv')
revenues_df = pd.read_csv('movie_revenues.csv')
tmdb_df = pd.read_csv('tmdb_movies.csv')

# Display the first few rows of each dataset
print("Movie Metadata:\n", metadata_df.head())
print("\nMovie Genres:\n", genres_df.head())
print("\nMovie Quotes:\n", quotes_df.head())
print("\nMovie Revenues:\n", revenues_df.head())
print("\nTMDB Movies:\n", tmdb_df.head())
```

```python
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect('movies.db')
cursor = conn.cursor()

# Insert data from DataFrames into the tables
metadata_df.to_sql('movie_metadata', conn, if_exists='replace', index=False)
genres_df.to_sql('movie_franchises', conn, if_exists='replace', index=False)
quotes_df.to_sql('movie_quotes', conn, if_exists='replace', index=False)
revenues_df.to_sql('movie_revenues', conn, if_exists='replace', index=False)
tmdb_df.to_sql('tmdb_movies', conn, if_exists='replace', index=False)

# Commit and close the connection
conn.commit()
conn.close()

print("Data successfully inserted into the database.")
```

```python
# Connect to the database
conn = sqlite3.connect('movies.db')

# Check inserted data
print(pd.read_sql_query("SELECT * FROM movie_metadata LIMIT 5;", conn))
print(pd.read_sql_query("SELECT * FROM movie_franchises LIMIT 5;", conn))
print(pd.read_sql_query("SELECT * FROM movie_quotes LIMIT 5;", conn))

conn.close()
```
